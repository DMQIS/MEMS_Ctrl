# Setup

#### Note about importing:
This connection setup uses the library `pyserial`.  This is NOT to be confused with the library `serial`.  It has the same name, but it's not what we're doing here.  To install the pyserial library, use the command `pip3 install pyserial`

If you're having library issues, try running `pip3 uninstall serial` and then reinstalling `pyserial`.

```python
from MEMS_serial_device import *
```

```python
port = "/dev/ttyUSB0"

Vbias = 90
HardwareFilterBW = 120
VdifferenceMax = 180
```

# Connect to device

Initialize serial connection to mirror & set mirror parameters

```python
## open a serial port
mems = MEMS_device(port=port)

## Prepare mirror settings
mems.set_mirror_params(Vbias=Vbias, HardwareFilterBW=HardwareFilterBW, VdifferenceMax=VdifferenceMax)
```

```python
## Print device parameters
print(mems)
```

# Turn on mirror

```python
mems.HV_on()
```

# Change position

```python
## Set the position
mems.set_mirror_position(x = 1, y = 1)
```

# Access information stored in class

```python
Vbias, Vdiff, HWfilt = mems.get_mirror_params(verbose=True)
x, y = mems.get_mirror_position(verbose=True)
```

# Disconnect without breaking stuff

```python
mems.exit_safely()
```
